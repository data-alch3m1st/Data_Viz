```python
# BUILDING THE RANDOM DATASET
# Define the number of rows
num_rows = 500

# Define datetime range
start_date = datetime(2024, 3, 15, 0, 5, 0)
end_date = datetime(2024, 3, 15, 15, 57, 0)
date_range = pd.date_range(start=start_date, end=end_date, periods=num_rows)

# Generate price step size with small random variations (to simulate sub-one-cent shytecoins;)
price_min = 0.000006300
price_max = 0.000007800
price_range = price_max - price_min
price_step = np.linspace(price_min, price_max, num_rows)

# Add small random variations to the price step
random_variations = np.random.uniform(-0.0000001, 0.0000001, num_rows)
price_step += random_variations

# Create the dataframe
df = pd.DataFrame({
    'datetime': date_range
    , 'price': price_step
    , 'amount': np.random.randint(
        10000, 825000001, size=num_rows)
    , 'exchange_name': np.random.choice(
        ['CoinThugEx', 'Crynance', 'BitBit', 'BitsOJoy', 'ScrumpEx']
        , size=num_rows, p=[0.05, 0.40, 0.25, 0.10, 0.20]
    )})

df.set_index('datetime', inplace=True)
df.sort_index(ascending=True, inplace=True)

print(df.shape)
print(df.info())
df.head()
```

```python
df['trade_amount'] = df['price'] * df['amount']
```

```python
print(df.shape)
print(df.info())
df.head()
```

```python
colorblind_palette = [
#     "#E69F00", 
    "#56B4E9", "#009E73", "#F0E442",
    "#0072B2", "#D55E00", "#CC79A7", "#999999",
    "#F28E2B", "#4E79A7"
]
```

```python

# my_palette = ['blue', 'red', 'green', 'purple', 'orange']

# Build the fig object

fig = px.scatter(
    df, x=df.index, y='price'
    , width=1200, height=1000
    , labels={
        'exchange_name': " Exchange"
        , 'datetime': " Trading Date/Time (GMT)"
        , 'price': " FAKE Price"
        , 'amount': " Amount (Tokens)"
    }
    , title="Trading in FAKE, the Ides of March"
    , color='exchange_name'
    , size='amount'
    , size_max=54
    , hover_data={'amount': ':,'}
    , color_discrete_sequence=colorblind_palette
)  # Use custom palette (e.g., BRGPO)

# Format the y-axis to adequate # of decimal places (e.g., 9 ~ likely high for 'sh!tcoins')

# Also include some format params for the cross-hair object;

fig.update_layout(
    yaxis_tickformat='.9f'
    , font_family="Menlo"
    , spikedistance=1
    , hoverdistance=1
)

fig.show();
```

```python
# Create custom palette (assign same # of colors as objects in the set assigned to {e.g. 'exchange_name'})

my_palette = ['blue', 'red', 'green', 'purple', 'orange']

# Build the fig object

fig4 = px.scatter(
    df, x=df.index, y='price'
    , width=1200, height=1000
    , labels={
        'exchange_name': " Exchange"
        , 'datetime': " Trading Date/Time (GMT)"
        , 'price': " FAKE Price"
        , 'amount': " Amount (Tokens)"
    }
    , title="Trading in FAKE, the Ides of March"
    , color='exchange_name'
    , size='amount', size_max=54
    , hover_data={'amount': ':,'}
    , color_discrete_sequence=my_palette
)  # Use custom palette (e.g., BRGPO)

# Format the y-axis to adequate # of decimal places (e.g., 9 ~ likely high for 'sh!tcoins')
# Also include some format params for the cross-hair object;

fig4.update_layout(yaxis_tickformat='.9f'
                   , font_family="Menlo"
                   , spikedistance=1
                   , hoverdistance=1
                  )

# Params for x-axis cross-hair object;

fig4.update_xaxes(
    showspikes=True   
    , spikecolor="green"    
    , spikesnap="cursor"          
    , spikemode="across")

# Params for y-axis cross-hair object;

fig4.update_yaxes(
    showspikes=True
    , spikecolor="gray"
    , spikethickness=2
    )

# Params for a Vertical Line (vline)
fig4.add_vline(
    x='2024-03-15 00:06:15.000000000'
    , line_width=3
    , line_dash="dash"
    , line_color="green"
)

# Params for a Horizontal Line (hline)
fig4.add_hline(
    y=0.000006650, line_dash="dash", line_color="green"
    , annotation_text="                    <<< Joe Trader Exit Point" 
    , annotation_position="top left"
)

# Show the fig!
fig4.show()
```
